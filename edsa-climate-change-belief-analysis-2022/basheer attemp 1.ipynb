## Data Cleaning 

Removing unnecessary noise e.g capitalization, symbols and stopwords 

```python
#replacing website limks with url-web if it exist in message in train data 
pattern_url = r'http[s]?://(?:[A-Za-z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9A-Fa-f][0-9A-Fa-f]))+'
subs_url = r'url-web'
df["message"] = df["message"].replace(to_replace = pattern_url, value = subs_url, regex = True)
```

```python
#replacing website limks with url-web if it exist in message in test data 
df_test["message"] = df_test["message"].replace(to_replace = pattern_url, value = subs_url, regex = True)
```

```python
# To avoid unneccessary noise, message column ws converted to lower case
df["message"]= df["message"]. str.lower()
df_test["message"]= df_test["message"].str.lower()
```

```python
#punctuation removal 
#function 
def remove_punctuation(post):
    return ''.join([l for l in post if l not in string.punctuation])
```

```python
df["message"]= df["message"].apply(remove_punctuation)
```

```python
df_test["message"]= df_test["message"].apply(remove_punctuation)
```

## Tokenization

```python
# sticking to use treebankwordtokenizer because it much quicker and easier
tokeniser = TreebankWordTokenizer()
df['message'] = df['message'].apply(tokeniser.tokenize)
```

```python
tokeniser = TreebankWordTokenizer()
df_test["message"]= df_test["message"].apply (tokeniser.tokenize)
```

### Lemmatization

```python
lemmatizer = WordNetLemmatizer()
```

```python
# building a function 
def mbti_lemma(words, lemmatizer):
    return [lemmatizer.lemmatize(word) for word in words]    
```

```python
# Applying the function on df and df_test (train and test dtata frame)
df['message'] = df['message'].apply(mbti_lemma, args=(lemmatizer, ))
```

```python
df_test["message"]= df ["message"].apply(mbti_lemma, args=(lemmatizer, ))
```
